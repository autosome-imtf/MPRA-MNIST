Currently, operations with genomic regions are implemented for Agarwal2025(AgarwalSingle, AgarwalMulti), Gosai2024, deAlmeida2022, Arensbergen2019, Kircher2019, Ernst2016 data.

# Agarwal dataset
Agarwal data uses hg38

```python
from mpramnist.Agarwal2025 import AgarwalSingleDataset, AgarwalMultiDataset
```

## Common use

```python
train = AgarwalSingleDataset(split="train", cell_type="HepG2", root="../data/")
val = AgarwalSingleDataset(split="val", cell_type="HepG2", root="../data/")
test = AgarwalSingleDataset(split="test", cell_type="HepG2", root="../data/")
print(len(train), len(val), len(test))
```

```python
train = AgarwalMultiDataset(split="train", cell_type=["HepG2", "K562", "WTC11"], root="../data/")
val = AgarwalMultiDataset(split="val", cell_type=["HepG2", "K562", "WTC11"], root="../data/")
test = AgarwalMultiDataset(split="test", cell_type=["HepG2", "K562", "WTC11"], root="../data/")
print(len(train), len(val), len(test))
```

## Exclude genomic regions

```python
test_regions = [
    {"chrom": "10", "start": 1, "end": 1000000000},
    {"chrom": "1", "start": 1, "end": 89032143},
]

test_dataset = AgarwalSingleDataset( # include to test dataset
    split="test", cell_type="WTC11", genomic_regions=test_regions, root="../data/"
)

train_dataset = AgarwalSingleDataset(split="train",cell_type="WTC11",
                                     genomic_regions=test_regions,
                                     exclude_regions=True, # exclude this from train
                                     root="../data/",
)
print(train_dataset)
print("===================")
print(test_dataset)
```

```python
test_dataset = AgarwalMultiDataset( # include to test dataset
    split="test", cell_type=["K562","HepG2","WTC11"], genomic_regions=test_regions, root="../data/"
)

train_dataset = AgarwalMultiDataset(split="train",cell_type=["K562","HepG2","WTC11"],
                                     genomic_regions=test_regions,
                                     exclude_regions=True, # exclude this from train
                                     root="../data/",
)
print(train_dataset)
print("===================")
print(test_dataset)
```

## Include genomic regions

```python
test_regions = [
    {"chrom": "10", "start": 1, "end": 1000000000},
    {"chrom": "1", "start": 1, "end": 1000000000},
    {"chrom": "20", "start": 1, "end": 1000000000},
]
val_regions = [
    {"chrom": "2", "start": 1, "end": 1000000000},
    {"chrom": "3", "start": 1, "end": 1000000000},
    {"chrom": "4", "start": 1, "end": 1000000000},
    {"chrom": "Y", "start": 1, "end": 1000000000},
]
train_regions = [
    {"chrom": "5", "start": 1, "end": 1000000000},
    {"chrom": "6", "start": 1, "end": 1000000000},
    {"chrom": "7", "start": 1, "end": 1000000000},
    {"chrom": "X", "start": 1, "end": 1000000000},
]

test = AgarwalSingleDataset(
    split="test", cell_type="K562", genomic_regions=test_regions, root="../data/"
)

val = AgarwalSingleDataset(
    split="val", cell_type="K562", genomic_regions=val_regions, root="../data/"
)

train = AgarwalSingleDataset(
    split="train", cell_type="K562", genomic_regions=train_regions, root="../data/"
)

print(len(train), len(val), len(test))
```

```python
test = AgarwalMultiDataset(
    split="test", cell_type=["K562","HepG2","WTC11"], genomic_regions=test_regions, root="../data/"
)

val = AgarwalMultiDataset(
    split="val", cell_type=["K562","HepG2","WTC11"], genomic_regions=val_regions, root="../data/"
)

train = AgarwalMultiDataset(
    split="train", cell_type=["K562","HepG2","WTC11"], genomic_regions=train_regions, root="../data/"
)

print(len(train), len(val), len(test))
```

## Download regions from BED-file

```python
test_regions = ["10\t1\t1000000000", "1\t1\t1000000000", "20\t1\t1000000000"]

with open("test_regions.bed", "w") as f:
    for region in test_regions:
        f.write(region + "\n")
```

```python
test_dataset = AgarwalSingleDataset(
    split="test",
    cell_type="HepG2",
    genomic_regions="./test_regions.bed",
    root="../data/",
)
train_dataset = AgarwalSingleDataset(
    split="train",
    cell_type="HepG2",
    genomic_regions="./test_regions.bed",
    exclude_regions=True,
    root="../data/",
)
print(train_dataset)
print("===================")
print(test_dataset)
```

```python
test_dataset = AgarwalMultiDataset(
    split="test",
    cell_type=["K562","HepG2","WTC11"],
    genomic_regions="./test_regions.bed",
    root="../data/",
)
train_dataset = AgarwalMultiDataset(
    split="train",
    cell_type=["K562","HepG2","WTC11"],
    genomic_regions="./test_regions.bed",
    exclude_regions=True,
    root="../data/",
)
print(train_dataset)
print("===================")
print(test_dataset)
```

# Gosai dataset
Gosai data uses hg19

```python
from mpramnist.Gosai2024 import GosaiDataset
```

```python
test_regions = [
    {"chrom": "10", "start": 1, "end": 1000000000},
    {"chrom": "1", "start": 1, "end": 1000000000},
    {"chrom": "X", "start": 1, "end": 1000000000},
]
train_dataset = GosaiDataset(
    split="train",
    filtration="none",
    root="../data/",
    genomic_regions=test_regions,
    exclude_regions=True,
)
test_dataset = GosaiDataset(
    split="test", filtration="none", root="../data/", genomic_regions=test_regions
)
print(train_dataset)
print("------------")
print(test_dataset)
```

```python
train_dataset = GosaiDataset(
    split="train",
    filtration="none",
    root="../data/",
    genomic_regions="./test_regions.bed",
    exclude_regions=True,
)
test_dataset = GosaiDataset(
    split="test",
    filtration="none",
    root="../data/",
    genomic_regions="./test_regions.bed",
)
print(train_dataset)
print("------------")
print(test_dataset)
```

# deAlmeida(DeepSTARR) dataset
deAlmeida data uses BDGP R5/dm3

```python
from mpramnist.deAlmeida2022 import deAlmeidaDataset
```

```python
val_regions = [
    {"chrom": "2R", "start": 1, "end": 21144049//2},
]
test_regions = [
    {"chrom": "2R", "start": 21144049//2, "end": 21144049},
]
val_test = val_regions + test_regions
train_dataset = deAlmeidaDataset(
    split="train",
    root="../data/",
    use_original_reverse_complement=True,
    genomic_regions=val_test,
    exclude_regions=True,
)
val_dataset = deAlmeidaDataset(
    split="val",  root="../data/", genomic_regions=val_regions
)
test_dataset = deAlmeidaDataset(
    split="test",  root="../data/", genomic_regions=test_regions
)
print(train_dataset)
print("===================")
print(val_dataset)
print("===================")
print(test_dataset)
```

# Arensbergen dataset
Arensbergen data uses hg19

```python
from mpramnist.Arensbergen2019 import ArensbergenDataset
print(ArensbergenDataset.GENOME_IDS)
```

```python
test_regions = [
    {"chrom": "10", "start": 1, "end": 1000000000},
    {"chrom": "1", "start": 1, "end": 1000000000},
    {"chrom": "X", "start": 1, "end": 1000000000},
]

task = "regression"

train_dataset = ArensbergenDataset(
    task=task,
    genome_id="SuRE42_HG02601",
    cell_type=['K562', 'HepG2'],
    split="train",
    root="../data/",
    genomic_regions=test_regions,
    exclude_regions=True,
) 

test_dataset = ArensbergenDataset(
    task=task,
    genome_id="SuRE42_HG02601",
    cell_type=['K562', 'HepG2'],
    split="test",
    root="../data/",
    genomic_regions=test_regions,
) 
print(train_dataset)
print("===================")
print(test_dataset)
```
